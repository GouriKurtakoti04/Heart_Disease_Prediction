```python
#import required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, accuracy_score, roc_auc_score,classification_report
```

```python
#Read Data
df=pd.read_csv("heart.csv")
df.head()
```

```python
#Outlier detection
num_cols = ['trestbps','chol','thalach','oldpeak','age']

#plot boxplots for each
plt.figure(figsize=(12,8))
for i, col in enumerate(num_cols):
    plt.subplot(2,3,i+1)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
#Outlier handling
def detect_outliers_iqr(data):
    outlier = []
    data = sorted(data)
    q1 = np.percentile(data,25)
    q3 = np.percentile(data,75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    return lower_bound,upper_bound

for col in num_cols:
    lower,upper = detect_outliers_iqr(df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
    df[col] = np.where(df[col] < lower, lower, df[col])
```

```python
#preprocessing
scaler = StandardScaler()
X = df.drop('target',axis=1)
y = df['target']
X_scaled = scaler.fit_transform(X)
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2,random_state=42)
```

```python
#Logistic Regression
Log_Reg = LogisticRegression()
Log_Reg.fit(X_train, y_train)
```

```python
#AUC score
y_prob = Log_Reg.predict_proba(X_test)[:, 1]
y_pred_log = Log_Reg.predict(X_test)

fpr,tpr,thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)
accuracy_log = accuracy_score(y_test, y_pred_log)
print("AUC Score: ",auc_score)
print("Accuracy: ",accuracy_log)
```

```python
#plot ROC Curve
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f'Model (AUC = {auc_score: .2f})')
plt.plot([0, 1], [0, 1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
#Confusion matrix
y_pred = Log_Reg.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: ")
print(conf_matrix)

#Visualizing with Heatmap
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['Healthy', 'Heart Disease'])
disp.plot(cmap='Blues')
plt.show()
```

```python
#KNN Classifier
knn = KNeighborsClassifier(n_neighbors=7)

#Training knn
knn.fit(X_train, y_train)
```

```python
#Predict output for X_test
y_predict = knn.predict(X_test)

#Calculate accuracy score
accuracy_score = accuracy_score(y_test, y_predict)
print("KNN Accuracy : ",accuracy_score)
```

```python
#Classification Report
class_report = classification_report(y_test,y_predict)
print("Classification Report:")
print(class_report)
```
